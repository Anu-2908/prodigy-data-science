```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style


time_data = {'whatapp':120, 'spotify': 90, 'Instagram': 40, 'Snapchat': 20}
social_media = list(time_data.keys())
time_spend = list(time_data.values())

plt.bar(social_media, time_spend)
plt.xlabel('Social Media')
plt.ylabel('Time in minutes')
plt.title('Average time spend on social media')

# applying different colours and size
style.use('ggplot')
color = {'royalblue', 'red', 'purple', 'yellow'}
plt.bar(social_media, time_spend, color=color, width=0.5)
plt.xlabel('Social Media')
plt.ylabel('Time in minutes')
plt.title('Average time spend on social media')
```

```python
fb = (120, 90, 110, 100, 95)
yt = (80, 90, 75, 85, 95)
ig = (50, 40, 55, 35, 45)
sc = (20, 30, 35, 25, 28)

xpos = np.arange(len(fb))
xpos
style.use('ggplot')
plt.figure(figsize=(10,7))
barwidth=0.2
plt.bar(xpos, fb, color='royalblue', width = barwidth, label='FB')

style.use('ggplot')
plt.figure(figsize=(10,7))
barwidth=0.2
plt.bar(xpos, fb, color='royalblue', width = barwidth, label='FB')
plt.bar(xpos, yt, color='red', width=barwidth, label="YT")
```

```python
# placing multiple graphs
style.use('ggplot')
plt.figure(figsize=(10,7))
barwidth=0.2
plt.bar(xpos, fb, color='royalblue', width = barwidth, label='FB')
plt.bar(xpos+0.2, yt, color='red', width=barwidth, label="YT")
plt.bar(xpos+0.4, ig, color='purple', width = barwidth, label='IG')
plt.bar(xpos+0.6, sc, color='yellow', width = barwidth, label='SC')
```

```python
# title with the legends
style.use('ggplot')
plt.figure(figsize=(10,7))
barwidth=0.2
plt.bar(xpos, fb, color='royalblue', width = barwidth, label='FB')
plt.bar(xpos+0.2, yt, color='red', width=barwidth, label="YT")
plt.bar(xpos+0.4, ig, color='purple', width = barwidth, label='IG')
plt.bar(xpos+0.6, sc, color='yellow', width = barwidth, label='SC')
plt.xticks(xpos+0.3, ('Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'))
plt.xlabel('Day')
plt.ylabel('Time spend in minutes')
plt.title('Time spend on social media')
plt.legend()
```

```python
# preparing a stacked bar chart
fb = (120, 90, 110, 100, 95)
yt = (80, 90, 75, 85, 95)
ig = (50, 40, 55, 35, 45)
sc = (20, 30, 35, 25, 28)
x = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
xpos = np.arange(len(x))
barwidth=0.4
plt.figure(figsize=(10,7))
plt.bar(xpos, fb, color='royalblue', width = barwidth, label='FB')
plt.bar(xpos, yt, color='red', width=barwidth, label="YT")
plt.bar(xpos, ig, color='purple', width = barwidth, label='IG')
plt.bar(xpos, sc, color='yellow', width = barwidth, label='SC')
plt.xlabel('Day')
plt.ylabel('Time spend in minutes')
plt.title('Time spend on social media')
plt.xticks(xpos, x)
plt.legend()
```
